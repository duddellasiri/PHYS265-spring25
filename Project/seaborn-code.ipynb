# Using seaborn to Analyze Airline Delay Trends

One way that seaborn extends matplotlib is providing a method to create violin plots which show the distribution of categorical data. In this notebook, I'm using a sample dataset of airline delay trends from January 2011 to December 2020 to depict the use of seaborn's violin plot feature

````python
import seaborn as sns

#print(sns.__version__)
%pip show seaborn
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Processing the data from the csv
df = pd.read_csv('delays_clean.csv')
df.head()
````

````python
fig, ax = plt.subplots(figsize=(14, 6)) #uses matplotlib to design figure
# clipped data to prevent extreme values from making the graph unreadable
sns.violinplot(data=df[df["arr_del15"] <= 300], x="carrier_name", y="arr_del15", inner="quartile", scale="width", ax=ax) #use of seaborn violinplot
ax.set_title("Arrival Delay Distributions by Airline (â‰¤ 300 mins)")
ax.set_xlabel("Carrier")
ax.set_ylabel("Arrival Delay (min)")
ax.tick_params(axis='x', rotation=45)

fig.tight_layout()
````

As shown in the above code block, seaborn can be used alongside its parent library, matplotlib to develop different types of plots using the same figure formatting techniques from matplotlib
